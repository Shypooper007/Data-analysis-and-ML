## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

fig, ax = plt.subplots(figsize = (10,10))

ax.plot(x,y)

ax.set(title='Simple Plot',xlabel='x-axis',ylabel='y-axis')

fig.savefig('images/sample-plot.png')
```

```python
car_sales['Total'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sale Date', y='Total')
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

```python
heart_disease.hist('age', bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots = True, figsize=(10,30));
```

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
len(over_50)
```

```python
## PYPLOT
over_50.plot(x='age',y='chol',kind = 'scatter', c='target', figsize=(10,6));
```

```python
## OO
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax = ax);
# ax.set_xlim([45,100])
```

```python
## IMP


fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(
       x = over_50['age'],
       y = over_50['chol'],
       c = over_50['target'])
ax.set(title='Heart Disease and Cholestrol Levels',
      xlabel='Age',
      ylabel='Cholestrol');
ax.legend(*scatter.legend_elements(),title='target')
ax.axhline(y = over_50['chol'].mean(),linestyle = '--')
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
scatter = ax1.scatter(
            x = over_50['age'],
            y = over_50['thalach'],
            c = over_50['target'])
ax1.set(xlabel = 'age',ylabel = 'cholestrol')
ax1.legend(*scatter.legend_elements(),title='target')
ax1.axhline(y = over_50['thalach'].mean(),linestyle = '--')
```
